```python
import csv
f = open('data/age_final.csv')
data = csv.reader(f)
result = []
for row in data: 
    if '신도림' in row[0]:
        for i in row[3:13]: ##3번째 데이터부터 12번째 데이터 까지 
              result.append(int(i))          
print(result)
```

```python
import csv
import matplotlib.pyplot as plt

f = open("data/age_final.csv")
data = csv.reader(f)
result = []

for row in data:
    if'후암동'in row[0]:
        for i in row[3:]:
            result.append(int(i))

plt.rc('font', family = "Malgun Gothic")
plt.title("후암동 인구분표")
plt.plot(result, "blue" ,label = "연령별분표")
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open("data/age_final.csv")
data = csv.reader(f)
result = []
t = input("사는 지역: ")

for row in data:
    if t in row[0]:
        for i in row[3:]:
            result.append(int(i))

plt.rc('font', family = "Malgun Gothic")
plt.title (t+" 인구분표")
plt.plot(result, "blue" ,label = "연령별분표")
plt.legend()
plt.show()
```
