```python
# Errors = “coerce” 라고 표기하면 숫자가 아닌 데이터를 NaN으로 설정하여 반환한다. 
# Pd.to_numeric 은 데이터를 숫자 형식으로 표출하라는 것. 

import pandas as pd
titanic = pd.read_csv("data/titanic.csv")

titanic["age"] = pd.to_numeric(titanic["age"], errors = "coerce")
```

```python
plt.hist(titanic["age"])
#다이타닉에 탑승한 승객들의 나이를 x축에 표기하고 y축은 승객의 수를 의마한다
```

```python
titanic["pclass"] = pd.to_numeric(titanic["pclass"], errors = "coerce")
plt.hist(titanic["pclass"])
```

```python
first_class = titanic[titanic["pclass"] ==1]["age"]
second_class =titanic[titanic["pclass"] ==2]["age"]
third_class = titanic[titanic["pclass"] ==3]["age"]

plt.figure(figsize = (10, 4))
ax = plt.subplot(1, 3, 1)
plt.hist(first_class, label = "1st Class", color= "#f1c40f", bins=20)
plt.title("1st Class")

plt.subplot(1,3,2,sharey=ax)
plt.hist(second_class, label = "1st Class", color= "#3498db", bins=20)
plt.title("2nd Class")

plt.subplot(1,3,3,sharey=ax)
plt.hist(third_class, label = "1st Class", color= "#e74c3c", bins=20)
plt.title("3rd Class")
```
