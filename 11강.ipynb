```python
died = [10,20,5,30,21]
germinated = [74,88,56,69,59]
```

```python
plants = ['Spinach','Turnip','Rhubarb','Broccoli','Kale']
died = [10,20,5,30,21]
germinated = [74,88,56,69,59]
```

```python
plt.figure(figsize = (6,6))
plt.bar(plants, germinated, color = 'orchid')
plt.title('Germination Rates')
plt.ylabel('Number Of Seeldings')
```

```python
plt.figure(figsize = (6,6))
plt.bar(plants, died, color = 'r')
plt.title('Failure Rates')
plt.ylabel('Number Of Seeldings')
```

```python
a = [10,25,5,30,21]
b = [74,88,56,69,59]
buttom = b
```

```python
plt.figure(figsize = (6,6))
plt.bar(plants, b, color = 'g', width = 0.4 ,align = "edge")
plt.bar(plants, a, color = 'b', width = 0.4)
```

```python
plt.figure(figsize = (6,6))
plt.bar(plants, b, color = 'g', label = 'Failed Seedlings')
plt.bar(plants, a, color = 'b', label = 'Germinated Seedlings', bottom=germinated)
plt.title('Greenhouse Seedlings')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (6,6))
plt.bar(plants, b, color = 'g', label = 'Failed Seedlings')
plt.bar(plants, a, color = 'b', label = 'Germinated Seedlings', bottom=germinated)
plt.legend(shadow = True, frameon = True, facecolor = 'white')
plt.title('Greenhouse Seedlings')
# plt.legend()
plt.show()
```

```python
# Errors = coerce      NaN  . 
# Pd.to_numeric      . 

import pandas as pd
titanic = pd.read_csv("data/titanic.csv")

titanic["age"] = pd.to_numeric(titanic["age"], errors = "coerce")
```

```python
plt.hist(titanic["age"])
#    x  y   
```

```python
titanic["pclass"] = pd.to_numeric(titanic["pclass"], errors = "coerce")
plt.hist(titanic["pclass"])
```

```python
plt.plot(heights, f_weights, marker = "*", label = "Female")
plt.plot(heights, m_weights, marker = "d", label = "Male")
plt.legend()
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
```

```python
first_class = titanic[titanic["pclass"] ==1]["age"]
second_class =titanic[titanic["pclass"] ==2]["age"]
third_class = titanic[titanic["pclass"] ==3]["age"]

plt.figure(figsize = (10, 4))
ax = plt.subplot(1, 3, 1)
plt.hist(first_class, label = "1st Class", color= "#f1c40f", bins=20)
plt.title("1st Class")

plt.subplot(1,3,2,sharey=ax)
plt.hist(second_class, label = "1st Class", color= "#3498db", bins=20)
plt.title("2nd Class")

plt.subplot(1,3,3,sharey=ax)
plt.hist(third_class, label = "1st Class", color= "#e74c3c", bins=20)
plt.title("3rd Class")
```
