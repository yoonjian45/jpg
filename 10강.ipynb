```python
import numpy as np
nums = np.arange(5)
plt.plot(nums,nums)
plt.plot(nums,nums*nums)
plt.plot(nums,nums**3)
plt.xlabel('time')
plt.ylabel('development')
plt.title('My development speed')
```

```python
import numpy as np
nums = np.arange(5)
plt.plot(nums,nums ,label = 'music')
plt.plot(nums,nums*nums, label = 'Math')
plt.plot(nums,nums**3 ,label = 'coding')
plt.xlabel('time')
plt.ylabel('development')
plt.title('My development speed')
plt.legend()
```
