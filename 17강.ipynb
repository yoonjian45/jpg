```python
def range(t):
    return t.max() -t.min()

titanic.groupby("pclass")["age"].agg(["min", "max", range])
```

```python
def range(s):
    return s.max() - s.min()


titanic.groupby("pclass")["age"].agg(["min", "max", range])
```
