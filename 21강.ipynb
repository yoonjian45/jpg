점심에는 남성과 여서의 식사 차이가 거의 없는데
저녁에는 여성보다 남성이 밖에서 더 많은 식사를 한다
남성은 점심보다 저녁에 다 많이 식사를 한다
여성은 점심과 저녁에 식사 차이가 거의 없다

점심과 저녁으로 나위었을 때, 남성은 점심보다 저녁에 돈을 더 많이쓰고,
여성은 점심과 저녁에 돈에 차이가 거의 없다.
넘성은 점심과 저녁에 흡연 비율이 비슷하고,
여성은 점심보다 저녁에 흡연자 수가 더 많다

```python
penguins = sns.load_dataset("penguins")
penguins
```

```python
sns.histplot(data=penguins, x="body_mass_g",hue = "species",bins=20,multiple = "stack")
```

```python
sns.histplot(data=penguins, x="body_mass_g",hue = "species",bins=20,multiple = "stack",kde = True)
plt.title("Penguin Body Mass")
```

```python
sns.set_theme()
plt.figure(dpi = 250)
sns.histplot(data=penguins, x="body_mass_g",hue = "species",bins=20,multiple = "stack",kde = True)
plt.title("Penguin Body Mass")
```
